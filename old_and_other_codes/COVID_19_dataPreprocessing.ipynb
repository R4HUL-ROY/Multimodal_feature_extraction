```python
import numpy as np
import pandas as pd
```

```python
filepath =  r"/content/drive/MyDrive/Machine Learning/ML DATASET/COVID-19 AR Clinical Correlates July202020 (csv).csv"
df = pd.read_csv(filepath, header = 1)
```

```python
df.head(5)
```

```python
print(df.shape)
df.drop(['RACE', 'ZIP', 'TEST NAME'], axis=1, inplace=True)
print(df.shape)
```

```python
df.set_index("PATIENT_ID", inplace = True)
df = df.iloc[:-2,:]  # remove last 2 NaN row

print(df.head())
print(df.shape)
print(df.tail(5))
```

```python
df.dropna(inplace=True)
print(df.shape)
```

```python
height_list = list(df["LATEST HEIGHT"])
print(height_list)
```

```python
## Here we converting the height from string format to cm
# 1 foot = 30.48 cm
# 1 inch = 2.54cm
import re
def heightStr_to_cm(height_str):
    number = re.split('"|\'| ',height_str)
    number = [float(i) for i in number if i != ""]
    return number[0]*30.48 + number[1]*2.54

df["LATEST HEIGHT"] = [heightStr_to_cm(i) for i in height_list]   
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
col_names = ['AGE', 'LATEST_BMI', 'LATEST WEIGHT', 'LATEST HEIGHT']
data = df[col_names]
print(data)

scaler = MinMaxScaler()
data = scaler.fit_transform(data)
df[col_names] = data
```
